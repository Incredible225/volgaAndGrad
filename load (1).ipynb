```python
xl1.columns=list(map(str.strip, xl1.columns))
xl2.columns=list(map(str.strip, xl2.columns))
xl3.columns=list(map(str.strip, xl3.columns))
xl4.columns=list(map(str.strip, xl4.columns))
xl5.columns=list(map(str.strip, xl5.columns))
xl6.columns=list(map(str.strip, xl6.columns))
```

```python
xl1=xl1.drop_duplicates()
xl2=xl2.drop_duplicates()
xl3=xl3.drop_duplicates()
xl4=xl4.drop_duplicates()
xl5=xl5.drop_duplicates()
xl6=xl6.drop_duplicates()
```

```python
colss=['hrId', 'name', 'type', 'ierId', 'level', 'point',
       'state', 'action', 'cardId', 'isNear', 'region', 'ierType', 'cityCode',
       'regionId', 'cityFiasId', 'isBlocking',
       'isMalicius', 'lostNumber', 'membership', 'streetCode', 'timeIsoStr',
       'isChemFlood', 'dispatchesId', 'districtCode', 'streetFiasId',
       'clarification', 'injuredNumber', 'responseStatus', 'actionTimeIsoStr',
       'assignTimeIsoStr',  'acceptOperatorStr']
```

```python
xl=pd.concat([xl1[colss],xl2[colss],xl3[colss],xl4[colss],xl5[colss],xl6[colss]],axis=0)
xl
```

```python
%%time

col=list(xl.select_dtypes(include=['object']).columns)
xl[col] = xl[col].apply(lambda x: x.str.strip())

```

```python
xl[['lat','lon']]=xl['point'].str.strip('[]{} ').str.replace('"lat":','').str.replace('"lon":','').str.replace(' ','').str.split(',',n=2, expand=True)
```

```python
xl=xl['type','state','action','isMalicius','membership','timeIsoStr','clarification','lat','lon']
```

```python
xl.to_pickle('my_data.pkl')
```
