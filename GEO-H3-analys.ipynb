```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
import h3
tqdm.pandas()
%matplotlib inline
```

```python
data = pd.read_pickle("./dataH3.pkl").reset_index(drop=True)
data
```

```python
rrr=data.groupby(by=['h3-10',"name"], dropna=False)['h3-10'].count()
rrr.index.names = ['h3','name']
rrr=rrr.reset_index()
rrr.columns=['h3','name','count']


```

```python
#h3.h3_to_geo_boundary(data_count['h3'].loc[i], True)[:6]
```

```python
top_data=pd.DataFrame()
for i in rrr['name'].unique():
    top_data=top_data.append(rrr[rrr['name']==i].sort_values(['count'],ascending=False)[:5])
top_data    
```

```python
top_data['lat']=0
top_data['lon']=0

for i in top_data.index:
    gps=h3.h3_to_geo_boundary(top_data['h3'].loc[i], True)
    top_data['lat'].loc[i]=(gps[0][0]+gps[3][0])/2
    top_data['lon'].loc[i]=(gps[0][1]+gps[3][1])/2
```
